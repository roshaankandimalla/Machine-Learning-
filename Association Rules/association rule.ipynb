```python
list=data.iloc[:,0].str.split(',').tolist()
list
```

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
encoder=TransactionEncoder()
finaldata=encoder.fit_transform(list)
finaldata
```

```python
finaldata_df=pd.DataFrame(finaldata,columns=encoder.get_feature_names_out())
```

```python
finaldata_df
```

```python
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
items=apriori(finaldata_df,min_support=0.05,use_colnames=True)
items
```

```python
rules = association_rules(items, metric="lift", min_threshold=1)
rules
```
