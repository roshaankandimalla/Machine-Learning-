```python
from keras.models import  Sequential  #Sequential
from keras.layers import Dense
from sklearn.model_selection import train_test_split
```

```python
x=finaldata.drop("letters",axis=1)
y=finaldata.letters
```

```python
y_encoded = pd.get_dummies(y)
```

```python
# 5ï¸âƒ£ Train-test split
x_train, x_test, y_train, y_test = train_test_split(
    x, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded
)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model=Sequential()
model.add(Dense(128,input_dim=number.shape[1],activation="relu"))
model.add(Dense(64,activation="elu"))
model.add(Dense(y_encoded.shape[1],activation="softmax"))
```

```python
model.compile(loss="categorical_crossentropy",optimizer="adam",metrics=["accuracy",'precision'])
```

```python
from scikeras.wrappers import KerasClassifier
```

```python
def create_model(n1=64,n2=32,activation='relu',optimizer='adam'):
    model=Sequential()
    model.add(Dense(n1,activation=activation,input_dim=x.shape[1]))
    model.add(Dense(n2,activation=activation))
    model.add(Dense(y.nunique(),activation='softmax'))
    if optimizer == 'adam':
        opt = Adam(learning_rate=0.01)
    elif optimizer == 'rmsprop':
        opt=RMSprop(learning_rate=0.001)
    else:
        opt=SGD(learning_rate=0.01)
    model.compile(optimizer=opt,loss='sparse_categorical_crossentropy',metrics=['accuracy'])
    return model
model=KerasClassifier(model=create_model,epochs=40,batch_size=64)
param_grid={
    'model__n1': [64, 128, 256],
    'model__n2': [32, 64, 128],
    'model__activation': ['relu', 'tanh'],
    'model__optimizer': ['adam', 'rmsprop', 'sgd']
}
```

```python
grid=RandomizedSearchCV(model,param_distributions=param_grid,n_iter=10,cv=3)
grid.fit(x_train,y_train)
```

```python
grid.best_params_
```
