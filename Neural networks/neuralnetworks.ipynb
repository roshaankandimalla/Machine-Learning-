```python
from scikeras.wrappers import KerasClassifier
```

```python
def create_model(n1=64,n2=32,activation='relu',optimizer='adam'):
    model=Sequential()
    model.add(Dense(n1,activation=activation,input_dim=x.shape[1]))
    model.add(Dense(n2,activation=activation))
    model.add(Dense(y.nunique(),activation='softmax'))
    if optimizer == 'adam':
        opt = Adam(learning_rate=0.01)
    elif optimizer == 'rmsprop':
        opt=RMSprop(learning_rate=0.001)
    else:
        opt=SGD(learning_rate=0.01)
    model.compile(optimizer=opt,loss='sparse_categorical_crossentropy',metrics=['accuracy'])
    return model
model=KerasClassifier(model=create_model,epochs=40,batch_size=64)
param_grid={
    'model__n1': [64, 128, 256],
    'model__n2': [32, 64, 128],
    'model__activation': ['relu', 'tanh'],
    'model__optimizer': ['adam', 'rmsprop', 'sgd']
}
```

```python
grid=RandomizedSearchCV(model,param_distributions=param_grid,n_iter=10,cv=3)
grid.fit(x_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

#  Discuss the performance differences between the model with default hyperparameters and the tuned model, emphasizing the effects of hyperparameter tuning.

Default Model (Before Tuning)

The initial model used default hyperparameters with 128 and 64 neurons in the two hidden layers.

It was trained using the Adam optimizer and categorical crossentropy as the loss function.

The model achieved 95% accuracy after 150 epochs.

Although the model performed well, it was not optimized for the dataset and required more epochs to converge.

3. Tuned Model (After Hyperparameter Tuning)

After hyperparameter tuning, the model architecture was modified to use 256 neurons in the first layer and 64 neurons in the second layer.

The loss function was changed to sparse categorical crossentropy, which is more suitable when using label-encoded target variables.

Despite reducing the training epochs from 150 to 40, the model achieved a higher accuracy of 96%.

The tuned model learned faster, generalized better, and showed improved efficiency compared to the default model.
