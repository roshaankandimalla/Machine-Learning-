```python
Q1={}
Q3={}
IQR={}
UE={}
LE={}
for i in data.columns :
 Q1[i]=data[i].quantile(0.25)
 Q3[i]=data[i].quantile(0.75)
 IQR[i]=Q3[i]-Q1[i]
 UE[i]=Q3[i]+1.5*IQR[i]
 LE[i]=Q1[i]-1.5*IQR[i]
 data.loc[data[i]>UE[i],i]=UE[i]
 data.loc[data[i]<LE[i],i]=LE[i]
```
